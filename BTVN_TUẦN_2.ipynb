```python
#Bài 2.11
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np

ride_distance = ctrl.Antecedent(np.arange(0, 100, 1), 'ride_distance')
traffic_condition = ctrl.Antecedent(np.arange(0, 101, 1), 'traffic_condition')
demand_level = ctrl.Antecedent(np.arange(0, 101, 1), 'demand_level')
weather_condition = ctrl.Antecedent(np.arange(0, 101, 1), 'weather_condition')
customer_rating = ctrl.Antecedent(np.arange(1, 6, 0.1), 'customer_rating')
ride_punctuality = ctrl.Antecedent(np.arange(0, 101, 1), 'ride_punctuality')

ride_distance['short'] = fuzz.trimf(ride_distance.universe, [0, 0, 3])
ride_distance['medium'] = fuzz.trimf(ride_distance.universe, [2, 5, 8])
ride_distance['long'] = fuzz.trimf(ride_distance.universe, [6,13,20])
ride_distance['very long'] = fuzz.trimf(ride_distance.universe, [15,50,50])

traffic_condition['low'] = fuzz.trimf(traffic_condition.universe, [0,0,30])
traffic_condition['medium'] = fuzz.trimf(traffic_condition.universe, [20,50,70])
traffic_condition['long'] = fuzz.trimf(traffic_condition.universe,[60,100,100])

demand_level['low'] = fuzz.trimf(demand_level.universe, [0,0,30])
demand_level['medium'] = fuzz.trimf(demand_level.universe, [20,50,70])
demand_level['high'] = fuzz.trimf(demand_level.universe, [60,100,100])

weather_condition['good'] = fuzz.trimf(weather_condition.universe, [0,0,40])
weather_condition['moderate'] = fuzz.trimf(weather_condition.universe, [20,35,50])
weather_condition['bad'] = fuzz.trimf(weather_condition.universe, [50,100,100])

customer_rating['poor'] = fuzz.trimf(customer_rating.universe, [1,1,2.5])
customer_rating['average'] = fuzz.trimf(customer_rating.universe, [2,3,4])
customer_rating['good'] = fuzz.trimf(customer_rating.universe, [3.5,5,5])

ride_punctuality['late'] = fuzz.trimf(ride_punctuality.universe, [0,0,50])
ride_punctuality['on_time'] = fuzz.trimf(ride_punctuality.universe, [40,60,80])
ride_punctuality['early'] = fuzz.trimf(ride_punctuality.universe, [70,100,100])

price = ctrl.Consequent(np.arange(0, 101, 1), 'price')
reward = ctrl.Consequent(np.arange(0, 101, 1), 'reward')

price['low'] = fuzz.trimf(price.universe, [0, 0, 30])
price['medium'] = fuzz.trimf(price.universe, [20, 50, 80])
price['high'] = fuzz.trimf(price.universe, [70, 100, 100])
price['very_high'] = fuzz.trimf(price.universe, [80, 100, 100])

reward['none'] = fuzz.trimf(reward.universe, [0, 0, 10])
reward['few'] = fuzz.trimf(reward.universe, [0, 15, 30])
reward['moderate'] = fuzz.trimf(reward.universe, [20, 50, 70])
reward['high'] = fuzz.trimf(reward.universe, [60, 100, 100])

rules = [
ctrl.Rule(ride_distance['short'] & traffic_condition['low'] & demand_level['low'], price['low']),
ctrl.Rule(ride_distance['short'] & traffic_condition['medium'] & demand_level['high'], price['medium']),
ctrl.Rule(ride_distance['medium'] & traffic_condition['long'] & demand_level['high'], price['high']),
ctrl.Rule(ride_distance['long'] & traffic_condition['medium'] & weather_condition['good'], price['medium']),
ctrl.Rule(ride_distance['long'] & traffic_condition['long'] & weather_condition['bad'], price['very_high']),
ctrl.Rule(ride_distance['very long'] & demand_level['high'], price['very_high']),
ctrl.Rule(ride_distance['medium'] & traffic_condition['low'] & demand_level['low'], price['medium']),
ctrl.Rule(ride_distance['short'] & traffic_condition['long'] & weather_condition['bad'], price['high']),
ctrl.Rule(ride_distance['very long'] & weather_condition['bad'], price['very_high']),
ctrl.Rule(ride_distance['medium'] & traffic_condition['medium'] & weather_condition['moderate'], price['medium']),


ctrl.Rule(customer_rating['good'] & ride_punctuality['early'], reward['high']),
ctrl.Rule(customer_rating['average'] & ride_punctuality['on_time'], reward['moderate']),
ctrl.Rule(customer_rating['poor'] & ride_punctuality['late'], reward['none']),
ctrl.Rule(ride_distance['long'] & traffic_condition['long'] & ride_punctuality['on_time'], reward['high']),
ctrl.Rule(ride_distance['medium'] & traffic_condition['medium'] & customer_rating['good'], reward['moderate']),
ctrl.Rule(customer_rating['poor'] & ride_punctuality['late'], reward['none']),
ctrl.Rule(ride_distance['very long'] & weather_condition['bad'] & customer_rating['good'], reward['high']),
ctrl.Rule(ride_distance['short'] & customer_rating['average'] & ride_punctuality['on_time'], reward['few']),
ctrl.Rule(ride_distance['long'] & traffic_condition['long'] & ride_punctuality['late'], reward['few']),
ctrl.Rule(ride_distance['medium'] & weather_condition['moderate'] & customer_rating['good'], reward['moderate'])
]

system = ctrl.ControlSystem(rules)
sim = ctrl.ControlSystemSimulation(system)

sim.input['ride_distance'] = 12
sim.input['traffic_condition'] = 80
sim.input['demand_level'] = 90
sim.input['weather_condition'] = 70
sim.input['customer_rating'] = 4.5
sim.input['ride_punctuality'] = 80
sim.compute()

print(sim.output['price'])
print(sim.output['reward'])
```

```python
#Bài 2.12
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

rating = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'rating')
sales_volume = ctrl.Antecedent(np.arange(0, 101, 1), 'sales_volume')
profit_margin = ctrl.Antecedent(np.arange(0, 51, 1), 'profit_margin')
seasonal = ctrl.Antecedent(np.arange(0, 101, 1), 'seasonal')
competitor = ctrl.Antecedent(np.arange(0, 71, 1), 'competitor')

discount = ctrl.Consequent(np.arange(0, 71, 1), 'discount')

rating['thap'] = fuzz.trimf(rating.universe, [0, 0, 4.0])
rating['trung_binh'] = fuzz.trimf(rating.universe, [4, 4.25, 4.5])
rating['cao'] = fuzz.trimf(rating.universe, [4.5, 5.0, 5.0])

sales_volume['thap'] = fuzz.trimf(sales_volume.universe, [0, 0, 40])
sales_volume['trung_binh'] = fuzz.trimf(sales_volume.universe, [30, 50, 70])
sales_volume['cao'] = fuzz.trimf(sales_volume.universe, [60, 100, 100])

profit_margin['thap'] = fuzz.trimf(profit_margin.universe, [0, 0, 15])
profit_margin['trung_binh'] = fuzz.trimf(profit_margin.universe, [10, 20, 30])
profit_margin['cao'] = fuzz.trimf(profit_margin.universe, [25, 40, 50])

seasonal['khong_co'] = fuzz.trimf(seasonal.universe, [0, 0, 30])
seasonal['trung_binh'] = fuzz.trimf(seasonal.universe, [20, 50, 70])
seasonal['cao'] = fuzz.trimf(seasonal.universe, [60, 100, 100])

competitor['thap'] = fuzz.trimf(competitor.universe, [0, 0, 10])
competitor['trung_binh'] = fuzz.trimf(competitor.universe, [5, 15, 25])
competitor['cao'] = fuzz.trimf(competitor.universe, [20, 40, 70])

discount['rat_thap'] = fuzz.trimf(discount.universe, [0, 2, 5])
discount['thap']     = fuzz.trimf(discount.universe, [5, 8, 10])
discount['trung_binh']= fuzz.trimf(discount.universe, [10, 15, 20])
discount['cao']      = fuzz.trimf(discount.universe, [20, 30, 40])
discount['rat_cao']  = fuzz.trimf(discount.universe, [40, 55, 70])

rules = [
ctrl.Rule(rating['cao'] & sales_volume['cao'] & profit_margin['cao'], discount['rat_thap']),
ctrl.Rule(rating['thap'] & sales_volume['thap'] & profit_margin['cao'], discount['cao']),
ctrl.Rule(seasonal['cao'] & competitor['cao'], discount['rat_cao']),
ctrl.Rule(rating['trung_binh'] & sales_volume['trung_binh'] & profit_margin['trung_binh'], discount['trung_binh']),
ctrl.Rule(competitor['thap'] & profit_margin['thap'] & sales_volume['cao'], discount['rat_thap']),
ctrl.Rule(rating['thap'] & seasonal['khong_co'], discount['trung_binh']),
ctrl.Rule(sales_volume['thap'], discount['cao'])
]

discount_ctrl = ctrl.ControlSystem(rules)
discount_sim = ctrl.ControlSystemSimulation(discount_ctrl)

def tinh_chiet_khau(rating_value, sales_value, profit_value, seasonal_value, competitor_value):
    discount_sim.input['rating'] = rating_value
    discount_sim.input['sales_volume'] = sales_value
    discount_sim.input['profit_margin'] = profit_value
    discount_sim.input['seasonal'] = seasonal_value
    discount_sim.input['competitor'] = competitor_value

    discount_sim.compute()
    return discount_sim.output['discount']

discount_sim.input['rating'] = rating_value
discount_sim.input['sales_volume'] = sales_value
discount_sim.input['profit_margin'] = profit_value
discount_sim.input['seasonal'] = seasonal_value
discount_sim.input['competitor'] = competitor_value

rating_value = 4.3
sales_value = 50
profit_value = 20
seasonal_value = 90
competitor_value = 50

discount_sim.compute()

print(discount_sim.output['discount'])
```

```python
#Bài 2.13
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

demand = ctrl.Antecedent(np.arange(0, 11, 1), 'demand')
competitor = ctrl.Antecedent(np.arange(0, 11, 1), 'competitor')
reputation = ctrl.Antecedent(np.arange(0, 11, 1), 'reputation')
profit = ctrl.Antecedent(np.arange(0, 11, 1), 'profit')
season = ctrl.Antecedent(np.arange(0, 11, 1), 'season')

demand['low'] = fuzz.trimf(demand.universe, [0, 0, 5])
demand['medium'] = fuzz.trimf(demand.universe, [2, 5, 8])
demand['high'] = fuzz.trimf(demand.universe, [5, 10, 10])
competitor['low'] = fuzz.trimf(competitor.universe, [0, 0, 5])
competitor['medium'] = fuzz.trimf(competitor.universe, [2, 5, 8])
competitor['high'] = fuzz.trimf(competitor.universe, [5, 10, 10])
reputation['low'] = fuzz.trimf(reputation.universe, [0, 0, 5])
reputation['medium'] = fuzz.trimf(reputation.universe, [2, 5, 8])
reputation['high'] = fuzz.trimf(reputation.universe, [5, 10, 10])
profit['low'] = fuzz.trimf(profit.universe, [0, 0, 5])
profit['medium'] = fuzz.trimf(profit.universe, [2, 5, 8])
profit['high'] = fuzz.trimf(profit.universe, [5, 10, 10])
season['low'] = fuzz.trimf(season.universe, [0, 0, 5])
season['medium'] = fuzz.trimf(season.universe, [2, 5, 8])
season['high'] = fuzz.trimf(season.universe, [5, 10, 10])

discount = ctrl.Consequent(np.arange(0, 71, 1), 'discount')

discount['very_low'] = fuzz.trimf(discount.universe, [0, 0, 5])
discount['low'] = fuzz.trimf(discount.universe, [5, 7, 10])
discount['medium'] = fuzz.trimf(discount.universe, [10, 15, 20])
discount['high'] = fuzz.trimf(discount.universe, [20, 30, 40])
discount['very_high'] = fuzz.trimf(discount.universe, [40, 70, 70])

rules = [
ctrl.Rule(demand['high'] & competitor['low'] & profit['low'],discount['very_low']),
ctrl.Rule(demand['low'] & competitor['high'] & profit['high'],discount['high']),
ctrl.Rule(reputation['high'] & profit['medium'] & season['high'],discount['medium']),
ctrl.Rule(competitor['high'] & season['high'] & profit['high'],discount['very_high']),
ctrl.Rule(reputation['low'] & demand['medium'] & profit['low'], discount['medium']),
ctrl.Rule(demand['high'] & season['low'] & competitor['low'],discount['very_low']),
ctrl.Rule(profit['high'] & competitor['medium'] & season['medium'],discount['medium']),
]
system = ctrl.ControlSystem(rules)
sim = ctrl.ControlSystemSimulation(system)
sim.input['demand'] = 9
sim.input['competitor'] = 5
sim.input['reputation'] = 4.2
sim.input['profit'] = 9
sim.input['season'] = 5

sim.compute()

print(sim.output['discount'])
```

```python
# Bài 2.14
!pip install scikit-fuzzy
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

order_density = ctrl.Antecedent(np.arange(0, 101, 1), 'order_density')
urgency = ctrl.Antecedent(np.arange(0, 101, 1), 'urgency')
driver_load = ctrl.Antecedent(np.arange(0, 101, 1), 'driver_load')
traffic = ctrl.Antecedent(np.arange(0, 101, 1), 'traffic')
profit_per_delivery = ctrl.Antecedent(np.arange(0, 101, 1), 'profit_per_delivery')

num_combine = ctrl.Consequent(np.arange(0, 101, 1), 'num_combine')
priority = ctrl.Consequent(np.arange(0, 101, 1), 'priority')

order_density['thap'] = fuzz.trimf(order_density.universe, [0, 0, 40])
order_density['trung_binh'] = fuzz.trimf(order_density.universe, [30, 50, 70])
order_density['cao'] = fuzz.trimf(order_density.universe, [60, 100, 100])

urgency['thap'] = fuzz.trimf(urgency.universe, [0, 0, 40])
urgency['trung_binh'] = fuzz.trimf(urgency.universe, [30, 50, 70])
urgency['cao'] = fuzz.trimf(urgency.universe, [60, 100, 100])

driver_load['thap'] = fuzz.trimf(driver_load.universe, [0, 0, 40])
driver_load['trung_binh'] = fuzz.trimf(driver_load.universe, [30, 50, 70])
driver_load['cao'] = fuzz.trimf(driver_load.universe, [60, 100, 100])

traffic['thap'] = fuzz.trimf(traffic.universe, [0, 0, 40])
traffic['trung_binh'] = fuzz.trimf(traffic.universe, [30, 50, 70])
traffic['cao'] = fuzz.trimf(traffic.universe, [60, 100, 100])

profit_per_delivery['thap'] = fuzz.trimf(profit_per_delivery.universe, [0, 0, 40])
profit_per_delivery['trung_binh'] = fuzz.trimf(profit_per_delivery.universe, [30, 50, 70])
profit_per_delivery['cao'] = fuzz.trimf(profit_per_delivery.universe, [60, 100, 100])

num_combine['it'] = fuzz.trimf(num_combine.universe, [0, 10, 30])
num_combine['mot_so'] = fuzz.trimf(num_combine.universe, [20, 40, 60])
num_combine['nhieu'] = fuzz.trimf(num_combine.universe, [50, 80, 100])

priority['thap'] = fuzz.trimf(priority.universe, [0, 10, 40])
priority['trung_binh'] = fuzz.trimf(priority.universe, [30, 50, 70])
priority['cao'] = fuzz.trimf(priority.universe, [60, 90, 100])

rules = [
ctrl.Rule(order_density['cao'] & driver_load['thap'] & traffic['thap'], num_combine['nhieu']),
ctrl.Rule(order_density['trung_binh'] & traffic['cao'] & urgency['trung_binh'], num_combine['mot_so']),
ctrl.Rule(driver_load['cao'] & order_density['cao'] & profit_per_delivery['trung_binh'], num_combine['mot_so']),
ctrl.Rule(order_density['thap'] & urgency['cao'] & traffic['trung_binh'], num_combine['mot_so']),
ctrl.Rule(profit_per_delivery['cao'] & urgency['cao'] & traffic['cao'], num_combine['mot_so']),
ctrl.Rule(urgency['cao'] & profit_per_delivery['cao'], priority['cao']),
ctrl.Rule(urgency['trung_binh'] & traffic['trung_binh'], priority['trung_binh']),
ctrl.Rule(urgency['thap'] & order_density['cao'] & profit_per_delivery['thap'], priority['thap'])
]

combine_system = ctrl.ControlSystem(rules)
combine_sim = ctrl.ControlSystemSimulation(combine_system)

def toi_uu_hoa_giao_hang(mat_do, khan_cap, tai_trong, giao_thong, loi_nhuan):
    combine_sim.input['order_density'] = mat_do
    combine_sim.input['urgency'] = khan_cap
    combine_sim.input['driver_load'] = tai_trong
    combine_sim.input['traffic'] = giao_thong
    combine_sim.input['profit_per_delivery'] = loi_nhuan
    combine_sim.compute()

    so_don_ket_hop = combine_sim.output['num_combine']
    muc_uu_tien = combine_sim.output['priority']

    return so_don_ket_hop, muc_uu_tien

mat_do = 80
khan_cap = 70
tai_trong = 20
giao_thong = 30
loi_nhuan = 70

so_don, uu_tien = toi_uu_hoa_giao_hang(mat_do, khan_cap, tai_trong, giao_thong, loi_nhuan)
print("Số lượng đơn hàng nên kết hợp :", round(so_don), "đơn (Nhiều)")
print("Mức độ ưu tiên giao hàng :", round(uu_tien), " (Trung bình)")

print("Hệ thống đề xuất: Nên kết hợp nhiều đơn hàng vì mật độ cao và tải trọng thấp")
print("Ưu tiên giao hàng ở mức trung bình")
```
